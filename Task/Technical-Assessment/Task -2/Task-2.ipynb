```python
import json
import pandas as pd
```

```python
# Opening JSON file
f = open('task_input_list.json')
  
# returns JSON object as 
# a dictionary
data = json.load(f)
```

```python
new=data['data']
for i in range(len(new)):
    print(new[i])
```

```python
amount=[]
for i in range(len(new)):
    try:
        if new[i][2]=='/':
            li=[]
            li.append(new[i])
            li.append(new[i+1])
            li.append(new[i+2])
            amount.append(li)
    except:
        print('###')
print(amount)
```

```python
deposit=[]
withdrawal=[]
for i in amount:
    am=i[2]
    da=i[0]
    if am[0]!='-':
        dep=[]
        dep.append(i[0])
        dep.append(i[1])
        dep.append(am)
        dep.append(da[0:2])
        dep.append(da[3:5])
        dep.append(da[6::])
        deposit.append(dep)
    else:
        wit=[]
        wit.append(i[0])
        wit.append(i[1])
        wit.append(am)
        wit.append(da[0:2])
        wit.append(da[3:5])
        wit.append(da[6::])
        withdrawal.append(wit) 
print(deposit)
print(withdrawal)
    
```

```python
depo=pd.DataFrame(deposit,columns=['Date','Description','Amount','month','day','year'])
withd=pd.DataFrame(withdrawal,columns=['Date','Description','Amount','month','month','year'])
```

```python
phone=[]
for i in new:
    if i.count('-')==3 and i.count(':')==1:
        a=i
        ph=''
        for j in a:
            if j.isnumeric() or (j=='-' or j=='(' or j==')'):
                ph=ph+j
        phone.append(ph)
    
    if i.count('-')>=1 and i.count('(')==1:
        
        a=i
        ph=''
        for j in a:
            if j.isnumeric() or (j=='-' or j=='(' or j==')'):
                ph=ph+j
        phone.append(ph)
print(phone)
```

```python
email:""
for i in new:
    a=i
    if "@" in a:
        if ".com" in a:
            email=email+i
            break
    else:
        email="NA"
email
```

```python
web=""
for i in new:
    a=i
    if ".com" in a:
        web=web+i
        break
web        
```

```python
#Converting Amount column form string to Float
depo['Amount'] = depo['Amount'].str.replace(',', '').astype(float)
withd['Amount'] = withd['Amount'].str.replace(',', '').astype(float)
```
